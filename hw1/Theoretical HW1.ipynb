```python
sum_abs_devs = lambda nums,nu: sum([*map(lambda z: abs(z-nu),nums)])
```

```python
sum_squared_dev = lambda nums,nu: sum([*map(lambda z: (z-nu)**2,nums)])
```

```python
NUM_COUNT = 11
MIN, MAX = -50, 50
```

```python
nums = np.random.randint(MIN,MAX,NUM_COUNT)
nums.sort(0)
sum_abs = lambda nu: sum_abs_devs(nums, nu)
sum_squared = lambda nu: sum_squared_dev(nums, nu)
mean_v, median_v = nums.mean(), median(nums)
print("Nums: %s" % nums)
print("Mean: %f, Sum: %f" % (mean_v,sum_abs(mean_v)))
print("Median: %f, Sum: %f" % (median_v,sum_abs(median_v)))
min_v, max_v = min([mean_v, median_v]),max([mean_v, median_v]),
X = np.linspace(-3 * abs(min_v), 3*abs(max_v),9*abs(max_v))

title("abs deviations")
sum_abs_vect = np.vectorize(sum_abs)
plot(X, sum_abs_vect(X));
figure()
title("squared deviations")
sum_sq_vect = np.vectorize(sum_squared)
plot(X, sum_sq_vect(X));
```
